```python
def contrast_stretching_gray(image):
    # Contrast Stretching
    # [0,80] => [0, 40]
    # [80, 220] => [40, 180]
    # [220, 255] => [180, 255]
    s1, s2 = 40, 180
    r1, r2 = 80, 220
    result = np.zeros_like(image, dtype=np.float32)
    alpha = s1 / r1
    beta = (s2 - s1) / (r2 - r1)
    gamma = (255 - s2) / (255 - r2)
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            r = image[i, j]
            if r < r1:
                s = alpha * r
            elif r < r2:
                s = s1 + (beta * (r - r1))
            else:
                s = s2 + (gamma * (r - r2))
            result[i, j] = s
    
    result = np.clip(result, 0, 255)  # Ensure values are within [0, 255]
    result = np.array(result, dtype=np.uint8)
    return result
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the image using matplotlib
image = plt.imread(r"C:\Users\Ruchit Kothari\OneDrive\Desktop\car Image 3.jpg")

# Display the original image
plt.imshow(image)
plt.show()

# Convert to grayscale
gray_scale = np.dot(image[...,:3], [0.2989, 0.587, 0.114]).astype(np.uint8)

# Display the grayscale image
plt.imshow(gray_scale, cmap='gray')
plt.title('Grayscaled Image')
plt.show()

# Contrast Stretching function
def contrast_stretching_gray(image):
    # Contrast Stretching
    # [0, 80] => [0, 40]
    # [80, 220] => [40, 180]
    # [220, 255] => [180, 255]
    s1, s2 = 40, 180
    r1, r2 = 80, 220
    result = np.zeros_like(image, dtype=np.float32)
    alpha = s1 / r1
    beta = (s2 - s1) / (r2 - r1)
    gamma = (255 - s2) / (255 - r2)
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            r = image[i, j]
            if r < r1:
                s = alpha * r
            elif r < r2:
                s = s1 + (beta * (r - r1))
            else:
                s = s2 + (gamma * (r - r2))
            result[i, j] = s

    result = np.clip(result, 0, 255)  # Ensure values are within [0, 255]
    result = np.array(result, dtype=np.uint8)
    return result

# Apply contrast stretching
stretched_image = contrast_stretching_gray(gray_scale)

# Display the contrast-stretched image
plt.imshow(stretched_image, cmap='gray')
plt.title('Contrast Stretched Image')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def high_pass_gray(image):
    kernel = np.array([
        [1/9, 1/9, 1/9],
        [1/9, -8/9, 1/9],
        [1/9, 1/9, 1/9]
    ])

    kernel_size = kernel.shape[0]
    padding = kernel_size // 2

    # Pad the image
    padded_image = np.pad(image, padding, mode='constant')

    # Initialize output image
    high_pass_image = np.zeros_like(image, dtype=np.uint8)

    # Apply the kernel
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            high_pass_image[i, j] = np.sum(kernel * padded_image[i:i+kernel_size, j:j+kernel_size])

    return high_pass_image

# Example usage with a given kernel
# Assuming gray_image_pika is your grayscale image array
result = high_pass_gray(gray_scale2.astype(np.uint8))
plt.imshow(result, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def low_pass_gray(image):
    kernel = np.array([
        [1/9, 1/9, 1/9],
        [1/9, 1/9, 1/9],
        [1/9, 1/9, 1/9]
    ])

    kernel_size = kernel.shape[0]
    padding = kernel_size // 2

    # Pad the image
    padded_image = np.pad(image, padding, mode='constant', constant_values=0)

    # Initialize output image
    low_pass_image = np.zeros_like(image, dtype=np.float32)

    # Apply the kernel
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            low_pass_image[i, j] = np.sum(kernel * padded_image[i:i+kernel_size, j:j+kernel_size])

    return low_pass_image

# Example usage with a simple averaging kernel
# Assuming gray_image_pika is your grayscale image array
result = low_pass_gray(gray_scale1)

# Display the resulting image using matplotlib
plt.imshow(result, cmap='gray')  # Show the image with grayscale colormap
plt.axis('off')  # Turn off axis labels
plt.show()  # Display the plot
```

```python
import numpy as np
import matplotlib.pyplot as plt

def high_boost_gray(image, scaling_factor):
    # Define the low-pass filter kernel (averaging filter)
    low_pass_kernel = np.array([
        [1/9, 1/9, 1/9],
        [1/9, 1/9, 1/9],
        [1/9, 1/9, 1/9]
    ])

    # Define the high-pass filter kernel
    high_pass_kernel = np.array([
        [-1/9, -1/9, -1/9],
        [-1/9, 8/9, -1/9],
        [-1/9, -1/9, -1/9]
    ])

    # Combine the kernels to form the high boost filter kernel
    kernel = low_pass_kernel + scaling_factor * high_pass_kernel

    kernel_size = kernel.shape[0]
    padding = kernel_size // 2

    # Pad the image
    padded_image = np.pad(image, padding, mode='constant', constant_values=0)

    # Initialize output image
    high_boost_image = np.zeros_like(image, dtype=np.float32)

    # Apply the kernel
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            high_boost_image[i, j] = np.sum(kernel * padded_image[i:i+kernel_size, j:j+kernel_size])

    return high_boost_image

# Example usage with a scaling factor of 2
# Assuming gray_image_pika is your grayscale image array
scaling_factor = 2
result = high_boost_gray(gray_scale1, scaling_factor)

# Display the resulting image using matplotlib
plt.imshow(result, cmap='gray')  # Show the image with grayscale colormap
plt.axis('off')  # Turn off axis labels
plt.show()  # Display the plot
```
